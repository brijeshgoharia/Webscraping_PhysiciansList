```python
#importing the libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url='https://www.asrs.org/find-a-specialist/results/List?page=0'

Name=[]
Number=[]
Street=[]
Locality=[]
Region=[]
Postal_code=[]
Country = []

for page in range(0, 45, 1):
    data = requests.get(url)
    soup = BeautifulSoup(data.text, 'html.parser')
    doc=soup.find("ol", {"class": "directory"})
    
    for a in doc.find_all('a', href=True):
        n =  ("https://www.asrs.org/"+a['href'])
        data1 = requests.get(n)
        soup = BeautifulSoup(data1.text, 'html.parser')
        #print(soup.prettify())
        #print(n)
        a = soup.get(n)
        #print(a)

        #print(soup.p)
        name=soup.find('h1')
        number=soup.find("p").contents[0].strip()
        streetAddress = soup.find('span', itemprop="streetAddress")
        streetAddress = streetAddress.text.replace('\n','')
        streetAddress = streetAddress.replace('\t','')   


        locality = soup.find('span', itemprop="addressLocality")
        #print(locality)
        region = soup.find('span', itemprop="addressRegion")
        postal = soup.find('span', itemprop="postalCode")
        #print(postal)
        country= soup.p.find('a').previousSibling.strip().split(" ")[1]

        Name.append(name.text)
        Number.append(number)
        Street.append(streetAddress.strip())
        Locality.append(locality.text)
        Region.append(region.text)
        Postal_code.append(postal.text)
        Country.append(country)

```

```python
df=pd.DataFrame({'Name':Name,'Number':Number,'Street':Street,'City':Locality, 'State':Region, 
                 'PostalCode':Postal_code, 'Country':Country})
```
